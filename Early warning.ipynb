```python
# 定义颜色转换函数
def color_change(val, min_val, max_val):
    color = 'red' if val < min_val or val > max_val else 'green'
    return f'color: {color}'

# 函数用于逐行处理DataFrame并应用样式
def apply_style(row):
    mean_h_style = color_change(row['Mean_H'], row['Hmin'], row['Hmax'])
    pred_h_style = color_change(row['Pred_H'], row['Hmin'], row['Hmax'])
    mean_r_style = color_change(row['Mean_R'], row['Rmin'], row['Rmax'])
    pred_r_style = color_change(row['Pred_R'], row['Rmin'], row['Rmax'])
    return [mean_h_style, pred_h_style,mean_r_style,pred_r_style]

# 使用Styler应用格式化
styled_df = merged[['Case','Time','Channel','Mean_H','Pred_H','Mean_R','Pred_R','Hmin','Hmax','Rmin','Rmax']].style.apply(
    lambda row: pd.Series(apply_style(row), index=['Mean_H', 'Pred_H','Mean_R','Pred_R']),
    axis=1
)

styled_df
```

```python
import pandas as pd
from colorama import Fore, Style, init

# 初始化colorama
init(autoreset=True)

def check_and_mark(value, min_value, max_value):
    if value < min_value or value > max_value:
        return f'Warning'
    else:
        return f'Normal'

# 使用apply方法更新DataFrame中的Mean_H和Pred_H列
merged['Warn_M_H'] = merged.apply(lambda row: check_and_mark(row['Mean_H'], row['Hmin'], row['Hmax']), axis=1)
merged['Warn_P_H'] = merged.apply(lambda row: check_and_mark(row['Pred_H'], row['Hmin'], row['Hmax']), axis=1)
merged['Warn_M_R'] = merged.apply(lambda row: check_and_mark(row['Mean_R'], row['Rmin'], row['Rmax']), axis=1)
merged['Warn_P_R'] = merged.apply(lambda row: check_and_mark(row['Pred_R'], row['Rmin'], row['Rmax']), axis=1)
merged
```

```python
filtered_df = merged[(merged['Warn_P_H'] == 'Warning') | (merged['Warn_P_R'] == 'Warning')]
filtered_df
```

```python
Sex_full = {'F':'Female','M':'Male'}
Time_full = {'a':'Afternoon','m':'Morning'}
Channel_full = {1:'C2-3',2:'C3-4',3:'C4-5',4:'C5-6',5:'C6-7',6:'C7-T1',7:'T1-2',8:'T2-3',9:'T3-4',10:'T4-5',11:'T5-6',12:'T6-7',13:'T7-8',14:'T8-9',15:'T9-10',16:'T10-11',17:'T11-12',18:'T12-L1',19:'L1-2',20:'L2-3',21:'L3-4',22:'L4-5',23:'L5-S1'}
```

```python
filtered_df['Mean_H'] = filtered_df['Mean_H'].round(2)
filtered_df['Pred_H'] = filtered_df['Pred_H'].round(2)
filtered_df['Mean_R'] = filtered_df['Mean_R'].round(2)
filtered_df['Pred_R'] = filtered_df['Pred_R'].round(2)
filtered_df['Sex'] = filtered_df['Sex'].map(Sex_full)
filtered_df['Time'] = filtered_df['Time'].map(Time_full)
filtered_df['Channel'] = filtered_df['Channel'].map(Channel_full)
```

```python
import cv2
import pandas as pd
import numpy as np
import os

output_folder = 'Warning_images'
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
    
def sharpen_image(image):
    # 定义锐化核
    kernel = np.array([
        [-1, -1, -1],
        [-1, 9, -1],
        [-1, -1, -1]
    ])
    # 使用卷积核进行锐化
    sharpened_image = cv2.filter2D(image, -1, kernel)
    return sharpened_image
    
def create_composite_image(row):
    # 读取图片
    composite_image = cv2.imread(row['IMG_File_Path'])

    # 调整图片的大小
    new_size = (512, 512)
    composite_image = cv2.resize(composite_image, new_size)
    composite_image = sharpen_image(composite_image)
    
    # 设置字体及字号
    font = cv2.FONT_HERSHEY_SIMPLEX
    font_scale = 0.8
    thickness = 2

    # 根据 Warn_M_H 列决定颜色
    if row['Warn_M_H'] == 'Warning':
        Warn_M_H_text_color = (0, 0, 255)  # 红色
    else:
        Warn_M_H_text_color = (0, 255, 0)  # 绿色

    if row['Warn_P_H'] == 'Warning':
        Warn_P_H_text_color = (0, 0, 255)  # 红色
    else:
        Warn_P_H_text_color = (0, 255, 0)  # 绿色
        
    if row['Warn_M_R'] == 'Warning':
        Warn_M_R_text_color = (0, 0, 255)  # 红色
    else:
        Warn_M_R_text_color = (0, 255, 0)  # 绿色
        
    if row['Warn_P_R'] == 'Warning':
        Warn_P_R_text_color = (0, 0, 255)  # 红色
    else:
        Warn_P_R_text_color = (0, 255, 0)  # 绿色
    
    # 在左上角标注 Mean_H, Pred_H, Mean_R, Pred_R 四列对应值
    cv2.putText(composite_image, f"Mean H: {row['Mean_H']}", (10, 20), font, font_scale, Warn_M_H_text_color, thickness, cv2.LINE_AA)
    cv2.putText(composite_image, f"Pred H: {row['Pred_H']}", (10, 45), font, font_scale, Warn_P_H_text_color, thickness, cv2.LINE_AA)
    cv2.putText(composite_image, f"Mean R: {row['Mean_R']}", (10, 70), font, font_scale, Warn_M_R_text_color, thickness, cv2.LINE_AA)
    cv2.putText(composite_image, f"Pred R: {row['Pred_R']}", (10, 95), font, font_scale, Warn_P_R_text_color, thickness, cv2.LINE_AA)

    # 在左下角标注 Sex, Time, Channel
    cv2.putText(composite_image, f"Sex: {row['Sex']}", (10, composite_image.shape[0] - 70), font, font_scale, (255, 255, 255), thickness, cv2.LINE_AA)
    cv2.putText(composite_image, f"Time: {row['Time']}", (10, composite_image.shape[0] - 45), font, font_scale, (255, 255, 255), thickness, cv2.LINE_AA)
    cv2.putText(composite_image, f"Channel: {row['Channel']}", (10, composite_image.shape[0] - 20), font, font_scale, (255, 255, 255), thickness, cv2.LINE_AA)

    # 保存图片
    composite_image_path = f"{output_folder}/composite_image_{row.name}.png"
    cv2.imwrite(composite_image_path, composite_image)
    print(f"Saved {composite_image_path}")

for index, row in filtered_df.iterrows():
    create_composite_image(row)
    #break  # 如果只需要处理一行，可以保留这个break，如果需要处理所有行，请移除这个break
```
